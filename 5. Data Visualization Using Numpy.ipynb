# Numpy Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(1000)
plt.hist(data, bins=30, alpha=0.5, color='blue')
plt.title("Histogram from Numpy Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

# Numpy Scatter Plot

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.title("Scatter Plot from Numpy Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Numpy Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title("Line Plot from Numpy Data")
plt.xlabel("X-axis")
plt.ylabel("Sine Values")
plt.show()
```
