# Seaborn

Seaborn is a powerful and user-friendly python library.

## Univariate Plots

These plot displays the distribution of a single variable.

### Histogram 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Generate random data
data=np.random.randn(1000)
```

```python
#Create a histogram
sns.histplot(data,bins=30,color='blue')
plt.title("Univariate plot - Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### KDE Plot (Kernel Density Estimate)

```python
sns.kdeplot(data,shade=True,color="red")
plt.title("Univariate plot - KDE")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

### Box Plot

```python
sns.boxplot(x=data,color="cyan")
plt.title("Univariate plot - Box Plot")
plt.xlabel("Values")
plt.show()
```

### Violin Plot 

```python
sns.violinplot(x=data,color="purple")
plt.title("Univariate plot - Violin Plot")
plt.xlabel("Value")
plt.show()
```

## Bivariate Plots 

These plots explore the relationship between two variables.

### Scatter Plots

```python
x=np.random.rand(100)
y=np.random.rand(100)
sns.scatterplot(x=x,y=y,color='green')
plt.title("Bivariate Plot - Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

### Line Plot

```python
x=np.linspace(0,10,100)
y=np.sin(x)
sns.lineplot(x=x,y=y,color="blue")
plt.title("Bivariate Plot - Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

### Heatmap 

```python
data=np.random.rand(10,10)
sns.heatmap(data,annot=True,cmap="coolwarm")
plt.title("Bivariate Plot - Heatmap")
plt.show()
```

### Bar Plot

```python
categories=["A","B","C"]
values=[10,20,15]
sns.barplot(x=categories,y=values,palette="muted")
plt.title("Bivariate plot - Barplot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

### Pair Plot(Special Bivariate Plot) 

```python
import pandas as pd
data=pd.DataFrame({"x":np.random.rand(50),
                  "y":np.random.rand(50),
                  "z":np.random.rand(50)})
sns.pairplot(data)
plt.suptitle("Bivariate Plot - PairPlot", y=1.02)
plt.show()
```
